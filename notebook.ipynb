## Data Validation
This dataset contains 15000 rows and 8 columns before cleaning and validation. All the columns have been validated against the criteria in the dataset table :
- week :- 6 distinct numeric values without missing values, cleaning not required.
- sales_method :- Replaced 'em + call' with 'Email + Call' and 'email' with 'Email'. 3 distinct text values after preprocessing.
- customer_id :- unique string values for each row without missing values, cleaning not required.
- nb_sold :- 10 distinct numeric values ranging from 7 to 16, cleaning not required.
- revenue :- numeric values, filled up the 'NA' values using Fill Up.
- years_as_customer :- numeric values ranging from 0 to 39, drop rows having value not in the range.
- nb_site_visits :- numeric values ranging from 15 to 41, cleaning not required.
